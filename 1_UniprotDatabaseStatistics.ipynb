```python
df['%'] =  ( df['Total'].astype(float) / float(df['Total'][:-1].sum() ) ) * 100
df['%'] = df['%'].apply(lambda x : np.round(x, 2))
```

```python
index = pd.MultiIndex.from_tuples(df.index)
df = df.replace(np.nan, 0)
df['SwissProt'] = df['SwissProt'].astype(int)
df['TREMBL'] = df['TREMBL'].astype(int)
df['Total'] = df['Total'].astype(int)

df = df.reindex(index)
```
